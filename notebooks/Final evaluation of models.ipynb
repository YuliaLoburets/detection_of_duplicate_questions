The comparison of models reveals a clear progression in performance:

Logistic Regression and Random Forest show modest results, with F1 scores around 0.70 and higher log loss values (>0.48).

XGBoost, built on a rich set of engineered features—including TF-IDF similarity, semantic similarity, embedding-based metrics, and lexical overlap—significantly improves performance, reaching an F1 score of 0.786 and reducing log loss to 0.341.

BERT stands out as the best-performing model with an F1 score of 0.880 and the lowest log loss (0.324), reflecting its superior ability to capture deep semantic relationships between questions.

Conclusion:
If computational resources allow, BERT is the preferred choice for this task. However, XGBoost, leveraging carefully engineered features, offers a strong and efficient alternative that balances performance and interpretability.

```python
fig, ax1 = plt.subplots(figsize=(10, 6))


ax1.bar(df['model'], df['f1_score'], color='skyblue', label='F1 Score')
ax1.set_ylabel('F1 Score', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(0.65, 0.95)


ax2 = ax1.twinx()
ax2.plot(df['model'], df['log_loss'], color='red', marker='o', label='Log Loss')
ax2.set_ylabel('Log Loss', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(0.3, 0.55)

plt.title('Model Comparison: F1 Score and Log Loss')
fig.tight_layout()
plt.grid(True)
plt.show()
```
